# Use BeautifulSoup4 to parse table from Wikipedia
# Use pandas to create a dataframe and pre-process

```python
with open('simple.html') as html_file:
    soup = BeautifulSoup(html_file, 'lxml')


source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text

soup = BeautifulSoup(source, 'lxml')
#print(soup.prettify())


Wiki_table = soup.find_all('table')[0]
Wiki_table_rows = Wiki_table.find_all('tr')

Code_list = []


for tr in Wiki_table_rows:
    td = tr.find_all('td')
    row = [tr.text for tr in td]
    if row:
        Code_list.append(row)
        

Code_table = pd.DataFrame(Code_list, columns = ['Postal Code','Borough','NeighborhoodOld'])
Code_table.replace(r'\n','',regex = True, inplace = True)
Code_table_new = Code_table[Code_table.Borough != 'Not assigned']
Code_table_new['Neighborhood']=Code_table_new['NeighborhoodOld'].replace('Not assigned',Code_table_new['Borough'])
Code_table_new2 = Code_table_new[['Postal Code','Borough','Neighborhood']]
Code_table_new2

Code_table_grouped = Code_table_new2.sort_values(['Neighborhood'], ascending = False).groupby(['Postal Code','Borough'])['Neighborhood'].apply(', '.join).reset_index()
Code_table_grouped


```

# get_category_type function

```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
    
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

# Clean up the json and structure it to pandas dataframe

```python
venues = results['response']['groups'][0]['items']

nearby_venues = json_normalize(venues) #flatten JSON

#filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_venues = nearby_venues.loc[:, filtered_columns]

# filter the category for each row
nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis = 1)

#clearn columns
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]

nearby_venues.head()
```

# Explore all neighborhoods

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

# Analyze Each Neighborhood

```python
# one hot encoding
Local_Toronto_onehot = pd.get_dummies(Local_Toronto_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
Local_Toronto_onehot['Neighborhood'] = Local_Toronto_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [Local_Toronto_onehot.columns[-1]] + list(Local_Toronto_onehot.columns[:-1])
Local_Toronto_onehot = Local_Toronto_onehot[fixed_columns]

Local_Toronto_onehot.head()
```

```python
Local_Toronto_onehot.shape
```

# group rows by neighborhood and by taking the mean of the frequency of occurence of each category

```python
Local_Toronto_grouped = Local_Toronto_onehot.groupby('Neighborhood').mean().reset_index()
Local_Toronto_grouped
```

```python
Local_Toronto_grouped.shape
```

# Print each neighborhood along with top 5 most common venues

```python
num_top_venues = 5

for hood in Local_Toronto_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = Local_Toronto_grouped[Local_Toronto_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

# Put the top venues in a pandas dataframe

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = Local_Toronto_grouped['Neighborhood']

for ind in np.arange(Local_Toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(Local_Toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted
```

# Cluster Neighborhoods

```python
# set number of clusters
kclusters = 5

Local_Toronto_grouped_clustering = Local_Toronto_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(Local_Toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

Create a new dataframe that includes the cluster as well as the top 10 venues for each neighborhood.

```python
Local_Toronto_merged = Local_Toronto_data

# add clustering labels
Local_Toronto_merged['Cluster Labels'] = kmeans.labels_

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
Local_Toronto_merged = Local_Toronto_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

Local_Toronto_merged # check the last columns!
```

# Visualize the clusters

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i+x+(i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(Local_Toronto_merged['Latitude'], Local_Toronto_merged['Longitude'], Local_Toronto_merged['Neighborhood'], Local_Toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
